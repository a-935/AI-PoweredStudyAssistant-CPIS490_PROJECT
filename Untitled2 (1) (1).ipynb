```python
import anthropic
# Initialize the Claude API client with your API key
CLAUDE_API_KEY = "API KEY..."

client = anthropic.Anthropic(api_key=CLAUDE_API_KEY)

print("✅ Claude is ready!")
```

```python
def summarize(text):
    message = client.messages.create(
        model="claude-haiku-4-5",
        max_tokens=1024,
        messages=[
            {
                "role": "user",
                "content": f"""You are a study assistant. Read the following study material and provide:
1. A concise summary (5-7 bullet points of the key ideas)
2. The most important concept in one sentence

Study material:
{text}"""
            }
        ]
    )
    return message.content[0].text


# ---- Test it ----
sample_text = """
Photosynthesis is the process by which green plants, algae, and some bacteria
convert light energy into chemical energy stored in glucose. It occurs mainly
in the chloroplasts, using chlorophyll to absorb sunlight. The process takes in
carbon dioxide and water, and releases oxygen as a byproduct. There are two main
stages: the light-dependent reactions and the Calvin cycle. Photosynthesis is
fundamental to life on Earth as it forms the base of most food chains.
"""

summary = summarize(sample_text)
print(summary)
```

```python
def generate_quiz(text):
    message = client.messages.create(
        model="claude-haiku-4-5",
        max_tokens=1024,
        messages=[
            {
                "role": "user",
                "content": f"""You are a study assistant. Based on the following study material, generate a quiz with:
1. 3 multiple choice questions (with 4 options each, indicate the correct answer)
2. 2 short answer questions

Study material:
{text}"""
            }
        ]
    )
    return message.content[0].text


# ---- Test it ----
quiz = generate_quiz(sample_text)
print(quiz)
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np

def generate_mindmap_image(topic, summary_text):
    # Use Claude to extract key concepts from the summary
    message = client.messages.create(
        model="claude-haiku-4-5",
        max_tokens=500,
        messages=[
            {
                "role": "user",
                "content": f"""Extract exactly 6 key concepts from this summary.
Return only a Python list of 6 short phrases (max 4 words each), nothing else.
Example format: ["Concept One", "Concept Two", "Concept Three", "Concept Four", "Concept Five", "Concept Six"]

Summary:
{summary_text}"""
            }
        ]
    )

    # Parse the concepts
    import ast
    concepts = ast.literal_eval(message.content[0].text.strip())

    # Draw the mind map
    fig, ax = plt.subplots(1, 1, figsize=(12, 8))
    ax.set_xlim(-3, 3)
    ax.set_ylim(-3, 3)
    ax.axis('off')
    fig.patch.set_facecolor('white')

    # Colors for branches
    colors = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4', '#FFEAA7', '#DDA0DD']

    # Center node
    ax.add_patch(mpatches.FancyBboxPatch((-0.6, -0.3), 1.2, 0.6,
        boxstyle="round,pad=0.1", facecolor='#1F4E79', edgecolor='white', linewidth=2))
    ax.text(0, 0, topic, ha='center', va='center', fontsize=11,
            fontweight='bold', color='white', wrap=True)

    # Branch positions around center
    angles = np.linspace(0, 2 * np.pi, len(concepts), endpoint=False)
    radius = 2.0

    for i, (concept, angle, color) in enumerate(zip(concepts, angles, colors)):
        x = radius * np.cos(angle)
        y = radius * np.sin(angle)

        # Draw line from center to concept
        ax.plot([0, x * 0.75], [0, y * 0.75], color=color, linewidth=2.5, alpha=0.7)

        # Draw concept box
        box_w, box_h = 1.0, 0.4
        ax.add_patch(mpatches.FancyBboxPatch(
            (x - box_w/2, y - box_h/2), box_w, box_h,
            boxstyle="round,pad=0.05", facecolor=color, edgecolor='white', linewidth=1.5, alpha=0.9))
        ax.text(x, y, concept, ha='center', va='center',
                fontsize=8, fontweight='bold', color='white',
                wrap=True, multialignment='center')

    plt.title(f"Mind Map: {topic}", fontsize=14, fontweight='bold', color='#1F4E79', pad=20)
    plt.tight_layout()
    plt.savefig(f"mindmap_{topic.replace(' ', '_')}.png", dpi=150, bbox_inches='tight', facecolor='white')
    plt.show()
    print(f"✅ Mind map generated!")
```

```python
def study_assistant(text, topic):
    print("=" * 60)
    print("📚 AI-POWERED STUDY ASSISTANT")
    print("=" * 60)

    print("\n🔹 STEP 1: GENERATING SUMMARY...\n")
    summary = summarize(text)
    print(summary)

    print("\n" + "=" * 60)
    print("\n🔹 STEP 2: GENERATING QUIZ...\n")
    quiz = generate_quiz(text)
    print(quiz)

    print("\n" + "=" * 60)
    print("\n🔹 STEP 3: GENERATING MIND MAP IMAGE...\n")
    generate_mindmap_image(topic, summary)  # ← passes summary now

    print("\n" + "=" * 60)
    print("✅ DONE!")
```

```python
from IPython.display import clear_output

# ---- User Input ----
print("📚 AI-POWERED STUDY ASSISTANT")
print("=" * 60)
topic = input("Enter the topic name (e.g. Photosynthesis): ")
text = input("Paste your study notes here: ")

clear_output()
study_assistant(text, topic)
```
